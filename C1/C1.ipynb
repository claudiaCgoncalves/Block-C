```python
# Importar Livrarias
import pandas as pd
import os
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
# Importar ficheiro csv
df = pd.read_csv('company_sales_data.csv')

# Criar Listas
profitList = df ['total_profit'].tolist()
monthList = df ['month_number'].tolist()
facecreamList = df ['facecream'].tolist()
facewashList = df ['facewash'].tolist()
toothpasteList = df ['toothpaste'].tolist()
bathingsoapList = df ['bathingsoap'].tolist()
shampooList = df ['shampoo'].tolist()
moisturizerList = df ['moisturizer'].tolist()

# Apresentar as 5 primeiras linhas
df.head()
```

```python
# Exercício I

# Definir títulos e eixos
plt.title('Company profit per month',fontsize=15)
plt.xlabel('Month Number',fontsize=15)
plt.ylabel('Profit in dollar',fontsize=15)

# Definir limites

plt.xlim(0, 13)
plt.ylim(100000, 500000)

# Definir frequência de números nas coordenadas ( 1 = min , 13 = max , 1 = freq)

plt.xticks(np.arange(1, 13, 1))
plt.yticks(np.arange(100000, 600000, 100000))

# Definir colunas a apresentar
plt.plot( monthList , profitList )

# Mostrar gráfico
plt.show()
```

```python
# Exercício II

# Definir variáveis

x1 = monthList
y1 = profitList

# Definir limites 

plt.xlim(0, 13)
plt.ylim(100000, 500000)
plt.xticks(np.arange(1, 13, 1))
plt.yticks(np.arange(100000, 600000, 100000))

# Definir Títulos

plt.title('Company profit per mont',fontsize=20)
plt.xlabel('Month Number',fontsize=15)
plt.ylabel('Profit in dollar',fontsize=15)

# Criar Gráfico 

plt.plot(x1, y1, marker='o', markerfacecolor='black', markersize=8, color='red', linewidth=3, linestyle='dashed')
plt.legend(["Profit data of last year"], loc="lower right")


plt.show()
```

```python
# Exercício III

# Definir variáveis para cada produto
x = monthList
y1 = facecreamList
y2 = facewashList
y3 = toothpasteList
y4 = bathingsoapList
y5 = shampooList
y6 = moisturizerList

# Definir títulos
plt.title('Sales data',fontsize=20)
plt.xlabel('Month Number',fontsize=15)
plt.ylabel('Sales units in number',fontsize=15)

# Definir limites
plt.xlim(0, 13)
plt.ylim(0, 18000)

# Criar o gráfico
plt.plot(x, y1, 'tab:blue', marker= 'o', label='Face Cream Sales Data')
plt.plot(x, y2, 'tab:orange', marker= 'o', label='Face Wash Sales Data')
plt.plot(x, y3, 'tab:green', marker= 'o', label='Tooth paste Sales Data')
plt.plot(x, y4, 'tab:red', marker= 'o', label='Bathing Soap Sales Data')
plt.plot(x, y5, 'tab:purple', marker= 'o', label='Shampoo Sales Data')
plt.plot(x, y6, 'tab:brown', marker= 'o', label='Moisturizer Sales Data')

# Posicionamento da legenda
plt.legend(loc='upper left')

plt.show()
```

```python
# Exercício IV

# Definir variáveis
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
totalunits = df["total_units"].sum()
facecream_data = df["facecream"].sum()
facewash_data = df["facewash"].sum()
toothpaste_data = df["toothpaste"].sum()
bathingsoap_data = df["bathingsoap"].sum()
shampoo_data = df["shampoo"].sum()
moisturizer_data = df["moisturizer"].sum()

# Criar o gráfico circular através da divisão entre número de produtos e total

Facecream = ((facecream_data) / (totalunits))
Facewash = ((facewash_data) / (totalunits))
Toothpaste = ((toothpaste_data) / (totalunits))
Bathingsoap = ((bathingsoap_data) / (totalunits))
Shampoo = ((shampoo_data) / (totalunits))
Moisturizer = ((moisturizer_data) / (totalunits))

exp_vals= [Facecream, Facewash, Toothpaste, Bathingsoap, Shampoo, Moisturizer]
exp_labels = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1, autopct=("%0.1f%%"))
plt.title('Sales Data') 


# Posicionamento da legenda
ax.legend(exp_labels,
    title="Sales Data",
    loc="lower right",
    bbox_to_anchor=(0.7, 0.1, 0.5, 0))

plt.show()
```
